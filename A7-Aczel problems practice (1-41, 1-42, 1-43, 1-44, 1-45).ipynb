#### 41. The following data are estimated worldwide appliance sales (in millions of dollars). Use the data to construct a pie chart for the worldwide appliance sales of the listed manufacturers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sales=[5100,4350,4180,3950,2200,2000,1580]
sales
```

```python
company=['Electrolux','General_Electric','Matsushita_Electric','Whirlpool','Bosch-Siemens','Philips','Maytag']
company
```

```python
plt.pie(sales, labels = company, autopct='%1.0f%%')
plt.show()
```
